```python
import os
import sys
from langchain.llms import Replicate
from langchain.vectorstores import Chroma
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import ConversationalRetrievalChain

# Replicate API token (if needed)
os.environ['REPLICATE_API_TOKEN'] = "r8_6QRkTGo0RElBPnmCytjLiXCtRDlSJRf0bLL8M"



# Update the pdf_path to the correct path in your Colab environment
pdf_path = '/content/docx.pdf'  # Adjust this path to match the file location in Colab

# Verify if the file exists before proceeding
if not os.path.isfile(pdf_path):
    raise ValueError(f"File path {pdf_path} is not a valid file.")

# Load and preprocess the PDF document using PyPDFLoader
loader = PyPDFLoader(pdf_path)
documents = loader.load()


# Use HuggingFace embeddings for transforming text into numerical vectors
embeddings = HuggingFaceEmbeddings()
vectordb=Chroma.from_documents(documents,embeddings)

# Initialize Replicate LLama2 Model
llm = Replicate(
    model="a16z-infra/llama13b-v2-chat:df7690f1994d94e96ad9d568eac121aecf50684a0b0963b25a41cc40061269e5",
    input={"temperature": 0.75, "max_length": 3000}
)

# Set up the Conversational Retrieval Chain
qa_chain = ConversationalRetrievalChain.from_llm(
    llm,
    vectordb.as_retriever(search_kwargs={'k': 2}),
    return_source_documents=True
)

# Start chatting with the chatbot
chat_history = []
while True:
    query = input('Prompt: ')
    if query.lower() in ["exit", "quit", "q"]:
        print('Exiting')
        sys.exit()
    result = qa_chain({'question': query, 'chat_history': chat_history})
    answer = result['answer']
    docs=result["source_documents"]
    print('Answer: ' + answer + '\n')
    print("surce docs:",docs)
    chat_history.append((query, answer))
```
